```python
data = "Ali-85,Sara-90,Ahmed-85,Zara-92,Ali-85"
print (data)
```

```python
records = data.split(",")
print(" Records List:", records)
```

```python
row = []
for record in records:
    name, marks = record.split("-")
    row.append((name, int(marks)))
print(" Tuples:", row)
```

```python
student_group = set(row)
```

```python
print("student group(remove duplicates name and marks):", student_group)
```

```python
student_dict = dict(student_group)
```

```python
print(" Dictionary:", student_dict)
```

```python
student_dict = {'Ali': 85, 'Sara': 90, 'Ahmed': 85, 'Zara': 92}
```

```python
high_name = ""
high_marks = 0
```

```python
for name, marks in student_dict.items():
    if marks > high_marks:
        high_marks = marks
        high_name = name
```

```python
print("highest marks:", (high_name, high_marks))
```

```python
names = tuple(student_dict.keys())
print("Tuple of student names:", names)
```

```python
uni_marks = set(student_dict.values())
```

```python
print(" count:", len(uni_marks))
```

```python
print("Unique marks are:", uni_marks)
```
